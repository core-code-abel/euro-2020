# Tabla de partidos

```python
import pandas as pd
import json

matches = pd.read_csv("./eurocup_2020_results.csv")
# df['match_id'] = [i + 1 for i in range(len(df))]
# df.head()
matches
```

## Eliminamos las columnas event_list, lineup_home y lineup_away
- La lista de eventos ir a otra tabla
- Las alineaciones irn (presumiblemente) a una tabla de partido/jugadores

```python
matches = matches.drop(columns=["events_list", "lineup_home", "lineup_away"])
```

```python
matches
```

## Limpieza de datos

### stage
- Final
- Semi-finals
- Quarter-finals
- Round of 16
- Group stage

```python
matches['stage'] = matches['stage'].apply(lambda x: 'Group stage' if 'Group stage' in x else x)
matches['stage'].value_counts()
```

### date
- La fecha viene en el formato DD.MM.YYYY y a postgres no le gusta...
- La pasamos a YYYY-MM-DD

```python
matches['date'] = matches['date'].apply(lambda x: '-'.join([item.strip() for item in x.split('.')[::-1]]))
matches['date'][0]
```

### Possession
- El valor viene como un string con el formato NN%.
- Lo pasamos a un float entre 0 y 1

```python
import re

def percentage_to_float(string):
    return int(re.findall(r'\d{1,2}', string)[0])/100

matches['possession_home'] = matches['possession_home'].apply(percentage_to_float)
matches['possession_away'] = matches['possession_away'].apply(percentage_to_float)
matches['possession_away'][0]
```

### Duels won
- El valor viene como un string con el formato NN%.
- Lo pasamos a un float entre 0 y 1

```python
matches['duels_won_home'] = matches['duels_won_home'].apply(percentage_to_float)
matches['duels_won_away'] = matches['duels_won_away'].apply(percentage_to_float)
matches['duels_won_away'][0]
```

### Pens score
- Cuando un partido no se decide por penaltys, el valor almacenado es False
- Lo sustituimos por NULL

```python
matches['pens_home_score'] = matches['pens_home_score'].apply(lambda x: 'NULL' if x == "False" else x)
matches['pens_away_score'] = matches['pens_away_score'].apply(lambda x: 'NULL' if x == "False" else x)
matches['pens_away_score'][0:3]
```

### Espacios en strings...
- Detectados espacios al principio o final de algunos elementos

```python
for col in matches.select_dtypes(include=['object']):
    matches[col]=matches[col].str.strip()
```

## Volcado a BBDD

###  Configuracin de conexin

```python
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv()

POSTGRES_USER = os.getenv('POSTGRES_USER')
POSTGRES_PASSWORD = os.getenv('POSTGRES_PASSWORD')
POSTGRES_HOST = os.getenv('POSTGRES_HOST')
POSTGRES_PORT = os.getenv('POSTGRES_PORT')
POSTGRES_DATABASE = os.getenv('POSTGRES_DATABASE')

url = f'postgresql+psycopg2://{POSTGRES_USER}:{POSTGRES_PASSWORD}@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DATABASE}'
print(url)
engine = create_engine(url)
connection = engine.connect()
```

### Creacin de la tabla partidos

```python

query = '''
    DROP TABLE IF EXISTS matches;
    CREATE TABLE matches (
        match_id serial PRIMARY KEY,
        stage VARCHAR(50) NOT NULL,
        date DATE NOT NULL,
        solve_on_pens BOOLEAN NOT NULL,
        home_match_team_id INTEGER NOT NULL,
        away_match_team_id INTEGER NOT NULL
    );
'''

connection.execute(query)
```

### Creacin de tabla partido/equipo

```python
query = '''
    DROP TABLE IF EXISTS match_team;
    CREATE TABLE match_team (
        match_id INTEGER NOT NULL,
        pens_score INTEGER,
        team_name VARCHAR(50) NOT NULL,
        team_score INTEGER NOT NULL,
        possession NUMERIC NOT NULL,
        total_shots INTEGER NOT NULL,
        shots_on_target INTEGER NOT NULL,
        duels_won NUMERIC NOT NULL
    );
'''

connection.execute(query)
```

### Recorrer el dataset para poblar las tablas

```python
query_matches = """
    INSERT INTO matches (
        stage,
        date,
        solve_on_pens,
        home_match_team_id,
        away_match_team_id
    )
    VALUES (
        '{stage}',
        '{date}',
        {pens},
        {home_match_team_id},
        {away_match_team_id}
    )
"""

query_match_team = """
    INSERT INTO match_team (
        pens_score,
        match_id,
        team_name,
        team_score,
        possession,
        total_shots,
        shots_on_target,
        duels_won
    )
    VALUES (
        {pens_score},
        {match_id},
        '{team_name}',
        {team_score},
        {possession},
        {total_shots},
        {shots_on_target},
        {duels_won}
    )
"""

for i, row in matches.iterrows():
    connection.execute(query_matches.format(
        stage = row["stage"],
        date = row["date"],
        pens = row["pens"],
        home_match_team_id = 2*i + 1,
        away_match_team_id = 2*i + 2,
    ))
    # TODO Refactor
    connection.execute(query_match_team.format(
        pens_score = row["pens_home_score"],
        match_id = i + 1,
        team_name = row["team_name_home"],
        team_score = row["team_home_score"],
        possession = row["possession_home"],
        total_shots = row["total_shots_home"],
        shots_on_target = row["shots_on_target_home"],
        duels_won = row["duels_won_home"]
    ))
    connection.execute(query_match_team.format(
        pens_score = row["pens_away_score"],
        match_id = i + 1,
        team_name = row["team_name_away"],
        team_score = row["team_away_score"],
        possession = row["possession_away"],
        total_shots = row["total_shots_away"],
        shots_on_target = row["shots_on_target_away"],
        duels_won = row["duels_won_away"]
    ))
```
